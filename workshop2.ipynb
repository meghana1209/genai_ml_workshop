```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

## sets up a preprocesing pipelines using columnTranfromer from scikit
## cate and numerical features

```python
numeric_transformer=StandardScaler()
oh_transformer=OneHotEncoder(drop='first') #as.type(float)
```

```python
#creating preprocessing pipelines

preprocessor=ColumnTransformer([("OneHotEncoder",oh_transformer,cat_features),
                             ("StandardScaler",numeric_transformer,num_features)])
```

```python
preprocessor
```

```python
X_train=preprocessor.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=preprocessor.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier #descision tree model creation and training
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,precision_score,recall_score
```

```python
###train your models
dt.fit(X_train,y_train)
lr.fit(X_train,y_train)
```
