```python
dummies = pd.get_dummies(df.town)  ### Converting categorical to numerical columns with binary values
dummies
```

```python
merged = pd.concat([df, dummies], axis = "columns")
```

```python
merged
```

```python
merged.drop(["town"], axis = "columns", inplace = True)
```

```python
merged
```

When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

```python
merged.drop(["west windsor"], axis = "columns", inplace = True)
```

```python
merged.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = merged.drop(["price"], axis = "columns") ## Input features data
```

```python
y = merged.price   #output variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test_pred = model.predict(X_test)
```

```python
y_test  #
```

```python
y_test_pred
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train, y_train)
```
